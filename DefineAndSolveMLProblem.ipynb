```python
# YOUR CODE HERE
# what we dont need after looking at df.head()
df = df.drop('country', axis = 1)
df = df.drop('Delivery Quality', axis = 1)
df = df.drop('Standard deviation of ladder by country-year', axis = 1)
df = df.drop('Standard deviation/Mean of ladder by country-year', axis = 1)
df = df.drop('GINI index (World Bank estimate), average 2000-15', axis = 1)
df = df.drop('gini of household income reported in Gallup, by wp5-year', axis = 1)

renaming = {'year': 'Year',
            'Life Ladder': 'Happiness', 
            'Log GDP per capita': 'Log GDP', 
            'Social support': 'Support', 
            'Healthy life expectancy at birth': 'Birth', 
            'Freedom to make life choices': 'Freedom',
            'Perceptions of corruption': 'Corruption', 
            'Confidence in national government': 'Confidence', 
            'GINI index (World Bank estimate)': 'GINI index',
            'Positive affect': 'Positive',
            'Negative affect': 'Negative',
            'Generosity': 'Generosity'}
df = df.rename(renaming, axis = 1)
```

```python
# YOUR CODE HERE
nan_count = np.sum(df.isnull(), axis = 0)
print(nan_count)
```

```python
# YOUR CODE HERE

# fill in missing NaNs
# PoC
df['Corruption_na'] = df['Corruption'].isnull()
df.loc[df['Corruption'].isnull()]
mean_PoC = df['Corruption'].mean()
df['Corruption'].fillna(value = mean_PoC, inplace = True)
# check if it worked
np.sum(df['Corruption'].isnull(), axis = 0)

# Positive affect
df['Positive_na'] = df['Positive'].isnull()
df.loc[df['Positive'].isnull()]
mean_plus = df['Positive'].mean()
df['Positive'].fillna(value = mean_plus, inplace = True)
# check if it worked
np.sum(df['Positive'].isnull(), axis = 0)

# Negative affect
df['Negative_na'] = df['Negative'].isnull()
df.loc[df['Negative'].isnull()]
mean_minus = df['Negative'].mean()
df['Negative'].fillna(value = mean_minus, inplace = True)
# check if it worked
np.sum(df['Negative'].isnull(), axis = 0)

# Log GDP per capita
df['Log GDP_na'] = df['Log GDP'].isnull()
df.loc[df['Log GDP'].isnull()]
mean_GDP = df['Log GDP'].mean()
df['Log GDP'].fillna(value = mean_GDP, inplace = True)
# check if it worked
np.sum(df['Log GDP'].isnull(), axis = 0)

# Social Support
df['Support_na'] = df['Support'].isnull()
df.loc[df['Support'].isnull()]
mean_SS = df['Support'].mean()
df['Support'].fillna(value = mean_SS, inplace = True)
# check if it worked
np.sum(df['Support'].isnull(), axis = 0)

# Healthy life expectancy at birth
df['Birth_na'] = df['Birth'].isnull()
df.loc[df['Birth'].isnull()]
mean_Birth = df['Birth'].mean()
df['Birth'].fillna(value = mean_Birth, inplace = True)
# check if it worked
np.sum(df['Birth'].isnull(), axis = 0)

# Freedom to make life choices
df['Freedom_na'] = df['Freedom'].isnull()
df.loc[df['Freedom'].isnull()]
mean_Free = df['Freedom'].mean()
df['Freedom'].fillna(value = mean_Free, inplace = True)
# check if it worked
np.sum(df['Freedom'].isnull(), axis = 0)

# Generosity
df['Generosity_na'] = df['Generosity'].isnull()
df.loc[df['Generosity'].isnull()]
mean_Gen = df['Generosity'].mean()
df['Generosity'].fillna(value = mean_Gen, inplace = True)
# check if it worked
np.sum(df['Generosity'].isnull(), axis = 0)

# Confidence in national government
df['Confidence_na'] = df['Confidence'].isnull()
df.loc[df['Confidence'].isnull()]
mean_Confidence = df['Confidence'].mean()
df['Confidence'].fillna(value = mean_Confidence, inplace = True)
# check if it worked
np.sum(df['Confidence'].isnull(), axis = 0)

# Democratic Quality
df['Democratic Quality_na'] = df['Democratic Quality'].isnull()
df.loc[df['Democratic Quality'].isnull()]
mean_Demo = df['Democratic Quality'].mean()
df['Democratic Quality'].fillna(value = mean_Demo, inplace = True)
# check if it worked
np.sum(df['Democratic Quality'].isnull(), axis = 0)

# GINI index (World Bank estimate)
df['GINI index_na'] = df['GINI index'].isnull()
df.loc[df['GINI index'].isnull()]
mean_GINI = df['GINI index'].mean()
df['GINI index'].fillna(value = mean_GINI, inplace = True)
# check if it worked
np.sum(df['GINI index'].isnull(), axis = 0)

# Remove na
df = df.drop(['Corruption_na',
              'Positive_na',
              'Negative_na',
              'Log GDP_na', 
              'Support_na', 
              'Birth_na', 
              'Freedom_na', 
              'Generosity_na', 
              'Confidence_na', 
              'Democratic Quality_na', 
              'GINI index_na'], axis = 1)

nan_count = np.sum(df.isnull(), axis = 0)
print(nan_count)
```

```python
# YOUR CODE HERE
y = df['Happiness']
X = df['Birth'].to_frame()

print(y.shape)
print(X.shape)

# create training and test data set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 42)
```

```python
# LinearRegression model object 
model = LinearRegression()

# fit the model
model.fit(X_train, y_train)

#  the predictions
prediction = model.predict(X_test)
```

```python
# evaluate
# mean squared error
print('\nModel Performance\n\nRMSE =   %.2f'
      % np.sqrt(mean_squared_error(y_test, prediction)))
# coefficient of determination --> 1 is perf
print(' R^2 =   %.2f'
      % r2_score(y_test, prediction))
```

```python
# visualize
plt.scatter(X.iloc[:, 0], y,  color = 'black', s = 15);

plt.plot(X_test, prediction, color = 'blue', linewidth = 3);

plt.xlabel('Birth');
plt.ylabel('Happiness');
```
